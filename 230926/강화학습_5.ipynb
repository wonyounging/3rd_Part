### cartpole 개요

```python
# 트랙에 cart 1개
# 1개의 pole
# cart에 +1 또는 -1의 힘을 주어 조절
# 막대기가 서있으면 time step마다 +1의 보상
# 종료 조건
#  막대기가 12도 이상 기울어짐
#  카트가 중심으로부터 2.4 이상 벗어남
#  time step > 500
#  100 번 시도에서 평균 195.0 이상의 점수를 얻으면 해결
```

```python
# 랜덤 액션

import warnings

warnings.filterwarnings("ignore")

from gymnasium.experimental.wrappers import RecordVideoV0
from IPython.display import Video
import gym

env = gym.make("CartPole-v1", render_mode='rgb_array')
env = RecordVideoV0(env, "video", name_prefix="cartpole1", disable_logger=True,episode_trigger=lambda x: x % 10 == 0 )
env.reset()

for _ in range(1000):
    env.render()
    env.step(env.action_space.sample())     # Take a random action
env.close()
```

```python
env = gym.make('CartPole-v1')

for i_episode in range(20):
  observation = env.reset()[0]
  for t in range(100):                        
      env.render()
      print(observation)
      action = env.action_space.sample()      
      observation, reward, done, info = env.step(action)[:4]
      if done:                                
          print('Episode finished after {} timesteps'.format(t+1))
          break

env.close()
```

```python
# 좌우 이동 : 막대가 오른쪽으로 기울어져 있으면 오른쪽으로 힘, 그렇지 않으면, 왼쪽으로 힘

env = gym.make('CartPole-v1', render_mode='rgb_array')

env = RecordVideoV0(env, "video", name_prefix="cartpole-method1")

observation = env.reset()[0]

for i in range(200):
    env.render()

    if observation[2] > 0:
        action = 1
    else: action = 0

    observation, reward, done, info = env.step(action)[:4]
    print(observation, done)

    # if done:
    #     print(i+1)
    #     break

env.close()
```
