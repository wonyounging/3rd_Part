### 동영상 캡처

```python
import cv2

import matplotlib.pyplot as plt

capture = cv2.VideoCapture("c:/workspace3/data/video/duck.mp4")

frames=int(capture.get(cv2.CAP_PROP_FRAME_COUNT))

frames

#현재 프레임 수(cv2.CAP_PROP_POS_FRAMES)

#영상의 총 프레임 수(cv2.CAP_PROP_FRAME_COUNT)
```

```python
plt.figure(figsize=(17,7))

i=0
cnt=int(frames/100)
frame_idx=0

while True:    
    ret, frame = capture.read()
   
    if ret:
        if i % 100 == 0:    
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            plt.subplot(2,3,frame_idx+1)
            plt.imshow(image)
            plt.axis("off")
            plt.title(f'frame {i+1}')

            frame_idx+=1

            if frame_idx>=cnt:
                break
    i += 1
    if i > frames:
        break

plt.show()    
```
