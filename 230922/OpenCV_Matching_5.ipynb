```python
# 템플릿 매칭 : 원본 이미지에서 템플릿 이미지와 일치하는 영역을 찾는 알고리즘

# 템플릿 이미지 조금씩 이동해가며 이미지 끝에 도달할 때까지 비교
# 템플릿 이미지와 동일하거나 가장 유사한 영역을 원본 이미지에서 검출

import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread("c:/workspace3/data/images/hats.png")
image_rgb1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread("c:/workspace3/data/images/hat.png")
image_rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb1)
plt.axis("off")
plt.show()

plt.imshow(image_rgb2)
plt.axis("off")
plt.show()
```
