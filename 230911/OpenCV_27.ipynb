### 키 이벤트 2

```python
import cv2

img_file = "c:/workspace3/data/images/pistol.jpg"
img = cv2.imread(img_file)
title = 'IMG'                  
x, y = 100, 100              

while True:
    cv2.imshow(title, img)
    cv2.moveWindow(title, x, y)

    key = cv2.waitKey(0) & 0xFF # 키입력 무한 대기
    print(key, chr(key))        

    if key == ord('h'):        
        x -= 10
    elif key == ord('j'):      
        y += 10
    elif key == ord('k'):      
        y -= 10
    elif key == ord('l'):      
        x += 10
    elif key == ord('q') or key == 27:
        break
      
    cv2.moveWindow(title, x, y )  

cv2.destroyAllWindows()
```
