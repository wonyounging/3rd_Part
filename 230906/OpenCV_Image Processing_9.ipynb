### 색상 반전

```python
#색상 반전: 어두운 부분은 밝게, 밝은 부분은 어둡게
# 255 - 픽셀값
# 200 => 255 - 200 = 55

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("c:/workspace3/data/images/plane.jpg")

image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis("off")

plt.show()
```

```python
out = img.copy()
print(out[0][0:5])

out = 255 - out
print(out[0][0:5])

plt.imshow(out)
plt.axis("off")

plt.show()
```
