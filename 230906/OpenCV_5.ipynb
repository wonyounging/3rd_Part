### 이미지-컬러채널

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("c:/workspace3/data/images/rear_garden.PNG")

print(img.shape)

(height, width) = img.shape[:2]
center = (width // 2, height // 2)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# 이미지의 채널 분할

(Blue, Green, Red) = cv2.split(img)
zeros = np.zeros(img.shape[:2], dtype = "uint8")

# red
img = cv2.cvtColor(cv2.merge([zeros, zeros, Red]), cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# green

img = cv2.cvtColor(cv2.merge([zeros, Green, zeros]), cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# blue

img = cv2.cvtColor(cv2.merge([Blue, zeros, zeros]), cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# 분할된 3채널 합치기

BGR = cv2.merge([Blue, Green, Red])

plt.imshow(BGR)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray,cmap='gray')
```
