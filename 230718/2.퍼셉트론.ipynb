```python
import matplotlib.pyplot as plt  
import numpy as np  
X=np.array([(0,0),(0,1),(1,0),(1,1)])  
y=np.array([0,0,0,1])  
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) #가중치
    b = -0.7 #절편(편향)
    tmp = np.sum(w*x) + b
    print(tmp)
    if tmp <= 0:
        return 0
    else:
        return 1
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = AND(xs[0], xs[1])
    print(xs , " -> " , y)
```

```python
y
```

```python
from sklearn.neural_network import MLPClassifier  
import mglearn 
y=[0,0,0,1]
model = MLPClassifier(random_state=0,max_iter=1000).fit(X, y)
mglearn.plots.plot_2d_separator(model, X, fill=True, alpha=.3)  
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("class 0")  
plt.ylabel("class 1")
```

```python
# NAND : AND의 역산(Not AND)
X=np.array([(0,0),(0,1),(1,0),(1,1)])
y=np.array([1,1,1,0])  
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5]) #and와 부호가 다름
    b = 0.7 #and와 부호가 다름
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = NAND(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

```python
y=[1,1,1,0]
model = MLPClassifier(random_state=0,max_iter=1000).fit(X, y)
mglearn.plots.plot_2d_separator(model, X, fill=True, alpha=.3)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("class 0")  
plt.ylabel("class 1")
```

```python
# OR
X=np.array([(0,0),(0,1),(1,0),(1,1)])
y=np.array([0,1,1,1])  
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2 #and와 값이 다름
    tmp = np.sum(w*x) + b
    print(tmp)
    if tmp <= 0:
        return 0
    else:
        return 1
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = OR(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

```python
y=[0,1,1,1]
model = MLPClassifier(random_state=0,max_iter=1000).fit(X, y)  
mglearn.plots.plot_2d_separator(model, X, fill=True, alpha=.3)  
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)  
plt.xlabel("class 0")  
plt.ylabel("class 1")
```

```python
#단층퍼셉트론의 한계 : xor를 풀 수는 없다. 하지만 게이트를 조합하면 가능하다.
X=np.array([(0,0),(0,1),(1,0),(1,1)])
y=np.array([0,1,1,0])  
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

```python
y=[0,1,1,0]
model = MLPClassifier(random_state=0,max_iter=1000).fit(X, y)  
mglearn.plots.plot_2d_separator(model, X, fill=True, alpha=.3)  
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("class 0")  
plt.ylabel("class 1")
```
