### 자르기

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("c:/workspace3/data/images/apple.jpg")
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(image_rgb.shape)

plt.imshow(image_rgb)
plt.title('original')
plt.axis("off")

plt.show()
```

```python
dst1 = image_rgb[100:500, 200:600].copy()
#                   행       열
dst2 = image_rgb.copy()

roi = image_rgb[100:300, 200:600]
# roi(Region of Interest) : 관심영역

dst2[0:200, 0:400] = roi
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,3,1)
plt.imshow(image_rgb)
plt.title('original')
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(dst1)
plt.title('dst1')
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(dst2)
plt.title('dst2')
plt.axis("off")

plt.show()
```
