### 산술연산 4

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("c:/workspace3/data/images/pencils.jpg")
src = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(src)
plt.axis("off")

plt.show()
```

```python
import numpy as np
import cv2

#연산을 위한 이미지

number = np.ones_like(src) * 127

_max = cv2.max(src, number)
_min = cv2.min(src, number)

#절대값의 차이
_abs = cv2.absdiff(src, number)

#비교 결과가 True이면 255, False이면 0
compare = cv2.compare(src, number, cv2.CMP_GT) #크면 255, 아니면 0
src = np.concatenate((src, src, src, src), axis = 1)
number = np.concatenate((number, number, number, number), axis = 1)

dst = np.concatenate((_max, _min, _abs, compare), axis = 1)
dst = np.concatenate((src, number, dst), axis = 0)
```

```python
plt.imshow(dst)
plt.title('image1')
plt.axis("off")

plt.show()
```
