### 트랜스폼

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("c:/workspace3/data/images/rear_garden.PNG")
print(img.shape)

(height, width) = img.shape[:2] # height,width
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
# (H, W, C)
```

```python
#이미지의 중심 좌표 구하기

(height, width) = img.shape[:2]
center = (width // 2, height // 2)

plt.imshow(img)

print("중심좌표:",center)
```

```python
# 이미지 회전

# 중심좌표, 각도(+:반시계방향, -:시계방향), 스케일
rotate = cv2.getRotationMatrix2D(center, -90, 1.5)  
rotated = cv2.warpAffine(img, rotate, (width, height))

plt.imshow(rotated)
```

```python
plt.imshow(img)
```

```python
# 이미지 resize

ratio = 400.0 / width
dimension = (400, int(height * ratio))
resized = cv2.resize(img, dimension, interpolation = cv2.INTER_AREA)
print(dimension)

plt.imshow(resized)
# x축의 범위 확인
```

```python
# 이미지 대칭

# Horizontal 1, Vertical 0, both -1

# flipped = cv2.flip(img, 0)  # 상하 대칭
# flipped = cv2.flip(img, 1)  # 좌우 대칭
flipped = cv2.flip(img, -1) # 원점 대칭

plt.imshow(flipped)
```
